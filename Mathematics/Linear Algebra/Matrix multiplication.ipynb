#### Matrix Multiplication

```python
import numpy as np
```

```python
A = np.array([
    [4,9,9],
    [9,1,6],
    [9,2,3]
])

B = np.array([
    [2,2],
    [5,7],
    [4,4]
])

print("Matrix of A: \n", A)
print("Matric of B: \n", B)
```

You can multiply matrices  ð´
  and  ðµ
  using NumPy package function np.matmul():

```python
np.matmul(A,B)
```

Which will output  3Ã—2
  matrix as a np.array. Python operator @ will also work here giving the same result:

```python
A @ B
```

#### Matrix Convention and Broadcasting

```python
try:
    np.matmul(B, A)
except ValueError as err:
    print(err)
```

```python
try :
    B@A
except ValueError as err:
    print(err)    
```

```python
x = np.array([1, -2, -5])
y = np.array([4, 3, -1])

print("Shape of vector x:", x.shape)
print("Number of dimensions of vector x:", x.ndim)
print("Shape of vector x, reshaped to a matrix:", x.reshape((3, 1)).shape)
print("Number of dimensions of vector x, reshaped to a matrix:", x.reshape((3, 1)).ndim)
```

```python
np.matmul(x,y)
```

```python
try:
    np.matmul(x.reshape((3,1)), y.reshape((3,1)))
except ValueError as err:
    print(err)    
```
